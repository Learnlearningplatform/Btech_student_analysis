## Methord 1:-
### "What is the number of students in each branch"

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.countplot(x='Branch',data=df)
plt.xlabel('Branch')
plt.ylabel('Number of students')
#ECE has the highest number of students
```

## Methord 2:-
### "Number of male and female in each branch"

```python
#number of male and female in each branch
df.groupby('Branch')['Gender'].value_counts()
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.countplot(x='Branch',data=df,hue='Gender')
plt.xlabel('Branch')
plt.ylabel('Number of students')
#cse has the highest number of girls
```

## Methord 3:-
### "Number of students of different category in each branch"

```python
df.groupby('Branch')['Category'].value_counts()
```

```python
fig,ax=plt.subplots(figsize=(15,5))
sns.countplot(x='Branch',data=df,hue='Category')
plt.xlabel('Branch')
plt.ylabel('Number of students')
```

## Methord 4:-
### "Percentage of each category in college"

```python
df['Category'].value_counts(normalize=True)
```

```python
m=df['Category'].value_counts(normalize=True)
m.plot.pie(figsize=(8,5))
```

## Methord 5:-
### "Average GPA of each branch"

```python
df.groupby('Branch')['GPA 1','GPA 2','GPA 3','GPA 4','GPA 5','GPA 6'].mean()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=3,figsize=(15,8))
plt.tight_layout()
sns.violinplot(x='Branch',y='GPA 1',data=df,ax=ax[0,0])
sns.violinplot(x='Branch',y='GPA 2',data=df,ax=ax[0,1])
sns.violinplot(x='Branch',y='GPA 3',data=df,ax=ax[0,2])
sns.boxplot(x='Branch',y='GPA 4',data=df,ax=ax[1,0])
sns.boxplot(x='Branch',y='GPA 5',data=df,ax=ax[1,1])
sns.boxplot(x='Branch',y='GPA 6',data=df,ax=ax[1,2])
```

## Methord 6:-
### "Maximum and Average CGPA of each branch"

```python
df.groupby('Branch')['CGPA'].agg(['max','mean']).sort_values(by='max',ascending=False)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(12,7))
plt.tight_layout()
sns.boxplot(x='Branch',y='CGPA',data=df,ax=ax[0])
ax[0].set_ylabel('AVERAGE CGPA OF BRANCH')
sns.barplot(x='Branch',y='CGPA',data=df,ax=ax[1],estimator=np.max)
ax[1].set_ylabel('Max cgpa of branch')
```
